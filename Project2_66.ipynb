```python
import pandas as pd
df_user = pd.read_table("/content/drive/MyDrive/python warehouse/users.dat", sep="::", header=None, names=['X'+str(i) for i in range(1,6)])
df_user
```

```python
df_rating = pd.read_table("/content/drive/MyDrive/python warehouse/ratings.dat", sep="::", header=None, names=['X'+str(i) for i in range(1,5)])
df_rating
```

```python
df_movies = pd.read_table("/content/drive/MyDrive/python warehouse/movies.dat", sep="::", header=None, names=['X'+str(i) for i in range(1,4)], encoding='ISO-8859-1')
df_movies
```

```python

merged_df = pd.concat([ df_movies,df_user,df_rating ], ignore_index=True,axis=1)
merged_df
```

```python

merged_df.to_hdf('data.h5', key='df', mode='w')
```
